```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# addition of sine waves to get the square wave

t = np.linspace(0,2,250)
f_not = 1
omega_not = 2 * np.pi * f_not
A = 5
x1 = A * np.sin(omega_not * t)
x2 = x1 + (A/3) * np.sin( 3* omega_not * t)
x3 = x2 + (A/5) * np.sin( 5 * omega_not *t)
x4 = x3 + (A/7) * np.sin( 7 * omega_not *t)
x5 = x4 + (A/9) * np.sin( 9 * omega_not *t)
x6 = x5 + (A/11) * np.sin( 11 * omega_not *t)

plt.plot(t,x6)
plt.grid()
```

```python
# square wave using a for loop

t = np.linspace(0,2, 250)
f_not = 1
omega_not = 2 * np.pi * f_not
A = 5

N = 25       # the harmonics to be included
xt = 0
for n in range(1, N+1, 2):
    xt = xt + (A/n)* np.sin( n * omega_not * t)

plt.plot(t,xt)
plt.grid()
```

```python
# Another example

t = np.linspace(0,2,300)
f_not = 1
omega_not = 2 * np.pi * f_not
A = 5

x0 = A/2
x1 = x0 + (A/np.pi)* np.sin(omega_not *t)
x2 = x1 - (A/2/np.pi)* np.sin(2*omega_not *t)
x3 = x2 + (A/3/np.pi)* np.sin(3*omega_not *t)
x4 = x3 - (A/4/np.pi)* np.sin(4*omega_not *t)
x5 = x4 + (A/5/np.pi)* np.sin(5*omega_not *t)

plt.plot(t,x5)
plt.grid()
```

```python
# Above example in a loop


t = np.linspace(0,2,300)
f_not = 1
omega_not = 2 * np.pi * f_not
A = 5
N = 25   # the harmonic    

xt = A/2

for n in range(1,N+1):
    xt = xt + ((-1)**(n+1)) * (A/n/np.pi)* np.sin(n*omega_not *t)

plt.plot(t,xt)
plt.grid()
```

```python
# The Impulse train 
# The Fourier coefficients do not decrease with n

t = np.linspace(0,2,500)
f_not = 1
omega_not = 2 * np.pi * f_not
A = 3
N = 30   # the harmonic    

xt = A

for n in range(1,N+1): 
    xt = xt + (2*A) * np.cos(n * omega_not *t)


plt.plot(t,xt)
plt.grid()
```

```python
# An Example: Weighted Summation of sinusoids
# Fourier series of triangular waveform 

t = np.linspace(0,2,300)
f_not = 1
omega_not = 2 * np.pi * f_not
N = 10             # Summation upto the Nth harmonic 
A = 6
xt = A*np.pi/2

for n in range(1,N+1,2):
    xt = xt - (A*2/np.pi/n/n)*np.cos(n*omega_not * t)
   
plt.plot(t,xt)
plt.grid()
```

```python
# Example: Weighted sum of sinusoids
# Example for Half-wave rectified waveform

t = np.linspace(0,2,300)
f = 1
A = 2
omega_not = 2 * np.pi * f
N = 11           # Summation to include Nth harmonic
xt = (A/np.pi) + (A/2)*np.sin(omega_not*t)

for n in range(2,N+1,2):        # even terms
    xt = xt - (2*A/np.pi/(n-1)/(n+1))* np.cos(n*omega_not * t)

plt.plot(t,xt)
plt.grid()
```

```python
# Example of a Full wave rectified waveform 


t = np.linspace(0,2,300)
f = 1
A = 3
omega_not = 2 * np.pi * f
N = 11
xt = (A/np.pi) 
for n in range(1,N+1):
    xt = xt + (A/np.pi)* (2/(1-4*n*n))* np.cos(n*omega_not * t)

plt.plot(t,xt)
plt.title('Series N terms ')
plt.legend()
plt.grid()

print('Number of terms in the series = ', N)
```
